```python
# 카드별로 사기거래 횟수를 알아보자.
fraud_card = cc_df2[cc_df2['is_fraud']== 1]['cc_num'].value_counts()
```

```python
# 카드별로 거래 횟수를 알아보자
total_card = cc_df2['cc_num'].value_counts()
```

```python
cc_df2['cc_num'].value_counts().sort_values(ascending=False)
# 눈에 띄는 한개가 있다.
# 사기거래가 가장 많은 카드번호와 거래가 가장많은 카드 번호와 똑같다.
```

```python
fraud_ratio = (fraud_card / total_card).fillna(0)
# nan값이 존재한다 > 사기거래가 없는 카드 0으로 채워주자!
# 하지만... 사기비율이 너무 낮다. 최고 0.4%?
```

# 가장 거리가 먼 거래내역 (번외)


```python
import folium
import numpy as np

#오 처음보는 함수... 혼자 이걸 어떻게 쓰지
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반경 (km)
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)

    a = np.sin(delta_phi / 2.0) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return R * c  # 거리 (km)


cc_df2["distance_km"] = haversine(cc_df2["lat"], cc_df2["long"], cc_df2["merch_lat"], cc_df2["merch_long"])
```

```python
# 가장 먼 거리의 결제 내역 찾기
max_distance_row = cc_df2.loc[cc_df2["distance_km"].idxmax()]

# 데이터 변환 (float 변환)
lat = float(max_distance_row["lat"])
long = float(max_distance_row["long"])
merch_lat = float(max_distance_row["merch_lat"])
merch_long = float(max_distance_row["merch_long"])

# 결제자 위치 및 결제 가게 위치 리스트 생성
cardholder_location = [lat, long]
merchant_location = [merch_lat, merch_long]

# 지도 생성 (결제자 위치 중심)
transaction_map = folium.Map(location=cardholder_location, zoom_start=8)

# 결제자 위치 마커 추가
folium.Marker(
    location=cardholder_location,
    popup="Cardholder Location",
    icon=folium.Icon(color="blue", icon="user")
).add_to(transaction_map)

# 결제가게 위치 마커 추가
folium.Marker(
    location=merchant_location,
    popup="Merchant Location",
    icon=folium.Icon(color="red", icon="shopping-cart")
).add_to(transaction_map)

# 결제자와 결제가게를 잇는 선 추가
folium.PolyLine(
    locations=[cardholder_location, merchant_location],
    color="black",
    weight=2.5,
    opacity=1
).add_to(transaction_map)

# 지도 출력
transaction_map

#시각화는 진짜 너무 어렵다
```
