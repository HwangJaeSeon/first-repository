```python
#3-1상관관계 밝히기
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])

data['date_diff']= data['tpep_dropoff_datetime'] - data['tpep_pickup_datetime']
data['date_diff']
```

```python
#주행시간과 요금
sns.scatterplot(x=data['date_diff'], y=data['fare_amount'])
```
